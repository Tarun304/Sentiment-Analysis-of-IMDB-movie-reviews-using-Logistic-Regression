Text Preprocessing

```python
#Lowercasing the review 
df['review']= df['review'].str.lower()
df
```

```python
#Removing the HTML tags
import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
df['review']=df['review'].apply(remove_html_tags)
df
```

```python
#Remove URLs
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
```

```python
df['review']=df['review'].apply(remove_url)
df
```

```python
import string
exclude=string.punctuation
exclude
```

```python
#Remove Punctuations
def remove_punc1(text):
    return text.translate(str.maketrans('', '', exclude))
```

```python
df['review']=df['review'].apply(remove_punc1)
df
```

```python
#Remove the stop words
from nltk.corpus import stopwords
```

```python
# Get the English stopwords as a set for faster membership checking
stop_words = set(stopwords.words('english'))

# Function to remove stop words from a review
def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word.lower() not in stop_words])

```

```python
df['review']=df['review'].apply(remove_stopwords)
df
```

```python
#Tokenization
import nltk
from nltk.tokenize import word_tokenize
df['review']=df['review'].apply(word_tokenize)
```

```python
import nltk
from nltk.corpus import wordnet as wn
from spacy.cli import download
from spacy import load
import warnings

nltk.download('omw-1.4')
nltk.download('wordnet')
nltk.download('wordnet2022')
nlp = load('en_core_web_sm')

! cp -rf /usr/share/nltk_data/corpora/wordnet2022 /usr/share/nltk_data/corpora/wordnet # temp fix for lookup error.



from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_tokens(tokens):
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return lemmatized_tokens
```

```python
# Apply lemmatization to the 'review' column
df['review'] = df['review'].apply(lemmatize_tokens)
df
```

```python
#TF-IDF representation
from sklearn.feature_extraction.text import TfidfVectorizer
df['review'] = df['review'].apply(' '.join) #Convert the review from a list to a string format
tfidf=TfidfVectorizer()
x=tfidf.fit_transform(df['review'])
```

```python
#Changing the sentiment values as 1 and 0 for positive and negative respectively
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(df['sentiment'])
```

```python
#Splitting the data into the training and the testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
#Building the Logistic Regression Model
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#Making the predictions
y_pred=model.predict(x_test)
```

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print("Accuracy:",accuracy*100, "%")
print("Precison:", precision)
print("Recall:", recall)
f1
```
